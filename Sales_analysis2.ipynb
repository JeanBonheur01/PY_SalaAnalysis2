## EXPLORATORY DATA ANALYSIS

### Q1: What is the overall sales trend monthly?

#### Step 1: extract the month in date and create a new column 

```python
#df["month_year"] = df["order_date"].apply(lambda x: x.strftime('%Y-%m'))

df['order_date'] = pd.to_datetime(df['order_date'])

# Extract year and month as 'month_year' in 'YYYY-MM' format
df['month_year'] = df['order_date'].dt.to_period('M').astype(str)
df["month_year"]
```

```python
df_monthly_trend= df.groupby("month_year")['sales'].sum().reset_index()
df_monthly_trend
```

```python
#Plotting 
plt.figure(figsize=(15, 6)) 
plt.plot(df_monthly_trend['month_year'], df_monthly_trend ['sales'])
plt.xlabel('Month-Year') 
plt.ylabel('Sales')      
plt.title('Monthly Sales Trend')  
plt.grid(True)  # Optional: add grid
plt.xticks(rotation='vertical', size = 8)  # Optional: rotate x-axis labels if needed
plt.tight_layout()  # Optional: adjust layout to fit labels
plt.show()
```

#### Finding 1: We observed that sales are initially low at the beginning of each year, but after four months, they start to rise significantly. Sales also decrease in July each year, but then begin to increase again, reaching a peak at the end of the year.

### Q2: What are the top 10 products by sales? 

```python
top_product_by_sales = df.groupby ("product_name")["sales"].sum()
sorted_top_product_by_sales = top_product_by_sales.sort_values(ascending=False).head(10)
sorted_top_product_by_sales
```

```python
# Plotting 

sorted_top_product_by_sales.plot(kind='bar', figsize=(10, 6), color='blue')

plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 
plt.show()
```

### Q3: What are the most selling products (Qt)? 

```python
top_selling_product= df.groupby("product_name")["quantity"].sum()
top_selling_product_sort= top_selling_product.sort_values(ascending=False).head(10)
top_selling_product_sort 
```

```python
# Plotting 

top_selling_product_sort.plot(kind= 'bar', figsize= (10, 6), color= 'orange')

plt.title('Top 10 Products by Quantity')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 
plt.show()
```

### Q4: What is the most prefered ship mode? 

```python
ship_mode= df.groupby("ship_mode").size()
ship_mode_sort= ship_mode.sort_values(ascending=True)
ship_mode_sort
```

```python
# Plotting 

ship_mode_sort.plot (kind= 'barh', figsize= (10, 6), color= 'green')

plt.ylabel ('Shipping mode')
plt.xlabel ('Count')
plt.title ('Distribution of prefered shipping mode')
plt.show
```

```python
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the count plot with different colors for each category
sns.countplot(x=df['ship_mode'], hue=df['ship_mode'], palette='viridis', dodge=False)
  # or use another palette like 'Set2', 'coolwarm', etc.

plt.title('Distribution of Preferred Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')

# Show the plot
plt.show()
```

### Q5: Which are the most profitable category and sub-category? 

```python
cat_subcat= df.groupby(['category', 'sub_category'])["profit"].sum().reset_index()
cat_subcat_sort= cat_subcat.sort_values(by=['category', 'sub_category'], ascending= False)
cat_subcat_sort
```

```python
category_profit= df.groupby("category")["profit"].sum()
category_profit_sort= category_profit.sort_values(ascending=True).reset_index()
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the count plot with different colors for each category
sns.barplot (x='category', y='profit', data= category_profit_sort, hue= 'category', palette='viridis', dodge= False)
  # or use another palette like 'Set2', 'coolwarm', etc.

plt.title('Total Profit by Category')
plt.xlabel('Categoty')
plt.ylabel('Total Profit')

# Show the plot
plt.show()
```

```python
sub_category_profit= df.groupby("sub_category")["profit"].sum()
sub_category_profit_sort= sub_category_profit.sort_values(ascending=True).reset_index()
# Set the figure size
plt.figure(figsize=(10, 6))

# Create the count plot with different colors for each category
sns.barplot (x='sub_category', y='profit', data= sub_category_profit_sort, hue= 'sub_category', palette='Set2', dodge= False)
  # or use another palette like 'Set2', 'coolwarm', etc.

plt.title('Total Profit by Sub_Category')
plt.xlabel('Sub_Categoty')
plt.ylabel('Total Profit')
plt.xticks(rotation=45, ha='right') 

# Show the plot
plt.show()
```
